```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
dias = 7  # Número de días para simular
pedidos_por_dia = 20  # Pedidos que recibe cada día
probabilidad_exito = 0.7  # Probabilidad de cumplir un pedido

# Generación de la variable aleatoria binomial para cada día de la semana
pedidos_satisfechos_semana = np.random.binomial(pedidos_por_dia, probabilidad_exito, dias)

# Impresión de resultados
print("Simulación de pedidos satisfechos por día durante una semana:")
for dia, pedidos_satisfechos in enumerate(pedidos_satisfechos_semana, start=1):
    print(f"Día {dia}: {pedidos_satisfechos} pedidos satisfechos")

# Generar la gráfica
dias_semana = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]

plt.figure(figsize=(10, 6))
plt.bar(dias_semana, pedidos_satisfechos_semana, color="skyblue")
plt.xlabel("Día de la Semana")
plt.ylabel("Pedidos Satisfechos")
plt.title("Pedidos Satisfechos por Día durante una Semana")
plt.ylim(0, pedidos_por_dia)  # Establece el límite máximo en la cantidad de pedidos por día
plt.show()

```
